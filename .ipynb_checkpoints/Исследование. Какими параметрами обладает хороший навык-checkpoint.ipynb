В рамках исследования рассмотрим различные фичи навыка(напр. категорию, ошибки, использование фичей платформы, разнообразнрость ответов и прочее) - по которым будем предсказывать различные аудиторные показатели навыка - напр. DAU, средняя оценка и т.д.

Если мы получим достаточно точную модель - то можем понять какие фичи вносят на текущий момент наибольший вклад в "успех" - т.о. получить алгоритм написания "хорошего" навыка

Стоит отдельно отметить, что очень важно следить за data leakage при построении модели - например при прогнозировании средней оценки нельзя использовать такую фичу как средняя длина сессии - так как длинные сессии косвенно означают хорошую оценку(будем это проверять с помощью корреляции) - словом нужно стараться использовать фичи которые есть только на этапе написания навыка - до того как его начали использовать юзеры.
